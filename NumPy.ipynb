```python
# 41. NumPy 배열의 기존 열에 새 열을 만들기
# iris_2d에 volume에 대한 새 열을 생성, 여기서 volume은 (pi x petallength x sepal_length^2)/3
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_2d = np.genfromtxt(url, delimiter=',', dtype='object')

sepallength = iris_2d[:, 0].astype('float')
petallength = iris_2d[:, 2].astype('float')
volume = (np.pi * petallength * (sepallength**2))/3

volume = volume[:, np.newaxis]

out = np.hstack([iris_2d, volume])
print(out[:4])
```

```python
# 49. 행 단위 계산
# 고유 값의 개수를 행 단위로 계산
np.random.seed(100)
arr = np.random.randint(1,11,size=(6, 10))

def counts_of_all_values_rowwise(arr2d):
    # 고유값 및 행 개수
    num_counts_array = [np.unique(row, return_counts=True) for row in arr2d]

    # 행 별 모든 값 카운트
    return([[int(b[a==i]) if i in a else 0 for i in np.unique(arr2d)] for a, b in num_counts_array])

print(arr, '\n')
print(np.arange(1,11), '<- key')
counts_of_all_values_rowwise(arr)
```

```python
# 51. 배열에 대한 원-핫 인코딩 생성
# 원-핫 인코딩(배열의 각 고유 값에 대한 더미 이진 변수)을 계산
np.random.seed(101) 
arr = np.random.randint(1,4, size=6)
print(arr, '\n')

def one_hot_encodings(arr): # 방법 1
    uniqs = np.unique(arr)
    out = np.zeros((arr.shape[0], uniqs.shape[0]))
    for i, k in enumerate(arr):
        out[i, k-1] = 1
    return out

print(one_hot_encodings(arr))

print((arr[:, None] == np.unique(arr)).view(np.int8)) # 방법 1
```

```python
# 67. numpy 배열의 이동 평균을 계산
np.random.seed(100)
Z = np.random.randint(10, size=10)
print('array: ', Z)

def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
  
print(moving_average(Z, n=3).round(2)) # 방법 1
print(np.convolve(Z, np.ones(3)/3, mode='valid')) # 방법 2
```

```python
# 69. 누락된 날짜를 채우기
dates = np.arange(np.datetime64('2022-04-01'), np.datetime64('2022-04-25'), 2)
print(dates)

filled_in = np.array([np.arange(date, (date+d)) for date, d in zip(dates, np.diff(dates))]).reshape(-1) # 방법 1
output = np.hstack([filled_in, dates[-1]])
print(output)

out = [] # 방법 2
for date, d in zip(dates, np.diff(dates)):
    out.append(np.arange(date, (date+d)))
filled_in = np.array(out).reshape(-1)
output = np.hstack([filled_in, dates[-1]])
print(output)
```

```python
# 70. 주어진 1D 배열에서 스트라이드를 생성
# 길이가 4이고 strides가 2
arr = np.arange(15) 

def gen_strides(a, stride_len=5, window_len=5):
    n_strides = ((a.size-window_len)//stride_len) + 1
    return np.array([a[s:(s+window_len)] for s in np.arange(0, n_strides*stride_len, stride_len)])

print(gen_strides(np.arange(15), stride_len=2, window_len=4))
```
